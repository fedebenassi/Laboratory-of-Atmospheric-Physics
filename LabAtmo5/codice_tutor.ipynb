```
#Program to read data from sonic anemometer
rm(list = ls())
library(pracma)

#setwd("C:/Users/Luigi/Desktop/Corso Lab")
mn <- readline(prompt = "Define the averaging window in seconds: ")
mm <- str2num(mn)


## Open and scan the dat file comma separated

file_in <- "TOA5_Sonic_Anemometer_G2.dat"

dati <- read.table(file_in, sep=",", na.string=NA, as.is=TRUE, header=TRUE, skip = 3)
options(digits.secs=3)
dati$X<-strptime(dati$X, '%Y-%m-%d %H:%M:%OS')

dati$date <- as.POSIXct(dati$X)

library(openair)

timePlot(dati, pollutant=c("Smp", "Smp.1", "Smp.2"),group = TRUE, name.pol=c("u","v","w"), ylab="(m/s)")

## Definition of the averaging window
npos <- as.numeric(nrow(dati))# data matrix dimension
samp_freq <- 20# sample frequency
n_int <- npos/(mm*samp_freq)# Number of intervals for the average windows
nA_avg <- npos/n_int# Elements of an averaging windows
nA_x <- 1:nA_avg


## Data
u <- dati[,3]# u-component of wind velocity (anemometer reference system --> positive toward north)
v <- dati[,4]# v-component of wind velocity (anemometer reference system --> positive toward west)
w <- dati[,5]# w-component of wind velocity (anemometer reference system --> positive upward)
t <- dati[,6]# sonic temperature
time <- 1:npos# time expressed in consecutive seconds from the first data


## Averaging
b <- 1

time_avg <- 0
uavg <- 0
vavg <- 0
wavg <- 0
tavg <- 0

for (bb in seq(0, npos-nA_avg, by = nA_avg)){
    time_avg[b] <- mean(time[bb+nA_x])# time in consecutive seconds from the first value
    uavg[b] <- mean(u[bb+nA_x])# running mean of u-component over the selected averaging window
    vavg[b] <- mean(v[bb+nA_x])# running mean of v-component over the selected averaging window
    wavg[b] <- mean(w[bb+nA_x])# running mean of w-component over the selected averaging window
    tavg[b] <- mean(t[bb+nA_x])# running mean of sonic temperature over the selected averaging window
    b <- b+1
    
}

time_avg <-time_avg[1:b]
uavg<-uavg[1:b]
vavg<-vavg[1:b]
wavg<-wavg[1:b]
tavg<-tavg[1:b]


output_avg2 <- cbind(time_avg, uavg, vavg, wavg, tavg)


write.table(output_avg2, file=paste(mn, "_sec_avg.dat"),  dec=".", sep ="\t", col.names=F, row.names=F, append=F)# averaged data are saved in the same folder


## Variance
var_u <- var(u)
var_v <- var(v)
var_w <- var(w)
var_t <- var(t)
readline(paste("Var(u) <- ", num2str(var_u)))
readline(paste("Var(v) <- ", num2str(var_v)))
readline(paste("Var(w) <- ", num2str(var_w)))
readline(paste("Var(t) <- ", num2str(var_t)))




## Probability Density Function
library(fitdistrplus)
pd_u <- plotdist(uavg, histo=TRUE)# histogram and cumulative density function of u
uavg <-as.vector(na.exclude(uavg))
pd_u <- fitdist(uavg, distr = "norm", method ="mle" )
denscomp(pd_u)
pd_v <- plotdist(vavg, histo=TRUE)# histogram and cumulative density function of v
vavg <-as.vector(na.exclude(vavg))
pd_v <- fitdist(vavg, distr = "norm", method ="mle" )
denscomp(pd_v)

pd_w <- plotdist(wavg, histo=TRUE)# histogram and cumulative density function of w
wavg <-as.vector(na.exclude(wavg))
pd_w <- fitdist(wavg, distr = "norm", method ="mle" )
denscomp(pd_w)

time_avg <-as.vector(na.exclude(time_avg))

plot(time_avg, uavg, ylim=c(0, max(wavg)), xlab = "time (s)", ylab = "intensity (m/s)", type="l", col="blue")

lines(time_avg,vavg, pch = 18, col = "red", type = "l", lty = 1)

lines(time_avg, wavg, pch = 18, col = "orange", lty = 1 )

legend("topright", legend=c("u", "v", "w"),col=c("blue", "red", "orange"), lty = 1, cex=0.8)
```
