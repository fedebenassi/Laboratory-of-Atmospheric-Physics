```
h = 0.180
delta_h = 0.006

V = 3.85
delta_V = 0.13

M_i = c(699, 725, 719, 630, 548)
M_f = c(553, 552, 627, 610, 542)

dM = 1

t = c(183, 124, 62, 255, 915)
delta_t = 1

N = c(24, 28, 15, 4, 1)
delta_N = 1

gocc = c('Doppio', 'Triplo', 'Rosso', 'Singolo', 'Ago')

M = M_i - M_f
delta_M = 2

M_star = N*V
delta_M_star = round((delta_V/V + delta_N/N)*M_star, digits = 0)

rate = round(N*h*3600/t, digits = 0)
delta_rate = round((delta_N/N + delta_h/h + delta_t/t)*rate, digits = 0)
```

```
df = data.frame(M = M, M_star = M_star, delta_M = delta_M, delta_M_star = delta_M_star, gocc = gocc, rate = rate, delta_rate = delta_rate)
df
```
